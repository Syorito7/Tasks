```python
import matplotlib.pyplot as plt
import numpy as np
import  scipy.integrate
def ode1(xs, t):
    x1, x2 = xs
    return [a1* x1*(1- b1*x2/k1), a2* x2*(1- b2*x1/k2)]
a1 = 1e-3
a2= 3e-3
b1= 6
k1= 1e-8
b2= 6
k2= 1e-8
t = 0
xmin = -4e-9
xmax= 2e-8
ymin = -4e-9
ymax = 2e-8
x = np.linspace(0, xmax, 40)
y = np.linspace(0, ymax, 40)
X, Y = np.meshgrid(x, y)

u, v = np.zeros(X.shape), np.zeros(Y.shape)

for i in range(40):
    for j in range(40):
        xy = ode1([X[i, j], Y[i, j]], t)
        u[i,j], v[i,j] = xy

plt.streamplot(X, Y, u, v, color='b',  density = 2,linewidth=0.5)
plt.plot(0, 0,'ro')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.xlim([xmin, xmax])
plt.ylim([ymin, ymax]);
```
